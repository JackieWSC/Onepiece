```python
from selenium import webdriver
```

```python
# setup the chome driver path and initial the chrome driver
chrome_driver = "/usr/local/bin/chromedriver"
driver = webdriver.Chrome(chrome_driver)
```

```python
# driver.get('https://www.lcsd.gov.hk/tc/leisurelink/index.html')
driver.get('http://leisurelink.lcsd.gov.hk/?lang=tc')
print("Access the page:{}".format(driver.title))
```

```python
# get text from html tag
actionBtnBlock = driver.find_element_by_css_selector('div [class="actionBtnBlock"]')
```

```python
if actionBtnBlock != None:
    print("Selected:{}".format(actionBtnBlock.text))
```

```python
actionBtnBlock.click()
```

```python
main_window = None
check_code_window = None
num_of_windows = len(driver.window_handles)

if num_of_windows > 1:
    main_window = driver.window_handles[0]
    check_code_window = driver.window_handles[1]

    driver.switch_to.window(check_code_window)
    print("New window:{}".format(driver.title))
```
